```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
x,y = make_regression(n_samples = 4, n_features = 1, n_informative = 1, n_targets = 1, noise = 80, random_state= 13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
#Lets apply ols
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x), color = 'red')
```

```python
#applying Gradient Descent assuming Slope m= 78.35
#and let's assume the starting value of intercept b=0

y_pred = ((78.35*x) + 0).reshape(4)
```

```python
plt.scatter(x,y)
plt.plot(x, reg.predict(x), color = 'red', label = 'ols')
plt.plot(x,y_pred, color = 'green', label = 'b=0')
plt.legend()
plt.show()
```

```python
m = 78.35
b = 0

loss_slope = -2 * np.sum(y - m*x.ravel() - b)
loss_slope
```

```python
#lets take the learning rate = 0.1

lr = 0.1

step_size = loss_slope *lr
step_size
```

```python
#calculate new intercept
b = b - step_size
b
```

```python
y_pred1 = ((78.35*x) + b).reshape(4)

plt.scatter(x,y)
plt.plot(x, reg.predict(x), color = 'black', label = 'ols')
plt.plot(x, y_pred1, color = 'blue', label = 'b = {}'.format(b))
plt.plot(x, y_pred, color = 'green', label = 'b = 0')
plt.legend()
```

```python
#iteration 2
loss_slope = -2*np.sum(y - m*x.ravel() - b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred2 = ((78.35 *x) + b).reshape(4)

plt.scatter(x,y)
plt.plot(x, reg.predict(x), color ='black', label = 'ols')
plt.plot(x, y_pred1, color = 'blue', label = 'b = 20.92 '.format(b))
plt.plot(x, y_pred2, color = 'red', label ='b = {}'.format(b))
plt.plot(x, y_pred, color = 'green', label = 'b = 0')

plt.legend()
plt.show()
```

```python
#iteration 3
loss_slope = -2 * np.sum(y - m*x.ravel() - b)
loss_slope
```

```python
step_size = loss_slope *lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred3 = ((78.35 *x) + b).reshape(4)
plt.figure(figsize = (15,15))

plt.scatter(x,y)
plt.plot(x, reg.predict(x), color ='black', label = 'ols')
plt.plot(x, y_pred1, color = 'blue', label = 'b = 20.92 '.format(b))
plt.plot(x, y_pred2, color = 'yellow', label = 'b = 25.11 '.format(b))
plt.plot(x, y_pred3, color = 'red', label ='b = {}'.format(b))
plt.plot(x, y_pred, color = 'green', label = 'b = 0')

plt.legend()
plt.show()
```

**Creating an iteration for Gradient Descend**

```python
b = -100
m = 78.35
lr = 0.1

epochs = 10

for i in range(epochs):
  loss_slope = -2 * np.sum(y - m*x.ravel() - b)
  b = b - (lr * loss_slope)

  y_pred = m *x +b

  plt.plot(x , y_pred)

plt.scatter(x,y)
```

```python
class GDRegressor:
  def __init__(self, learning_rate, epochs):
    self.m = 95.98
    self.b = -120
    self.lr = learning_rate
    self.epochs = epochs

  def fit(self, x,y):
    #calculate the b using GD

    for i in range(self.epochs):
      loss_slope = -2 * np.sum(y - self.m*x.ravel() - self.b)
      self.b = self.b - (self.lr* loss_slope)
      print(loss_slope, self.b)
    print(self.b)


```
